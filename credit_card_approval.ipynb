```
#converting ? to nan
import numpy as np
credit_card=credit_card.replace('?',np.nan)
credit_card.tail(20)
```

```
credit_card.isnull().sum()
```

```
#fill the missing value using most frequent data in that particular column
for feature in credit_card:
  if credit_card[feature].dtype=='object':
    credit_card=credit_card.fillna(credit_card[feature].value_counts().index[0])
```

```
credit_card.isnull().sum()
```

```
#converting categorical data to numerical data
from sklearn.preprocessing import LabelEncoder

label_encoder=LabelEncoder()

for feature in credit_card:
  if credit_card[feature].dtype=='object':
    credit_card[feature]=label_encoder.fit_transform(credit_card[feature])
```

```
credit_card.head()
```

```
#splitting data
from sklearn.model_selection import train_test_split
credit_card=credit_card.drop([11,13],axis=1)
credit_card=credit_card.values
X,Y=credit_card[:,0:13],credit_card[:,-1]
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=50)
```

```
#scalling the data between o and 1
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(0,1))
rescaled_Xtrain=scaler.fit_transform(X_train)
rescaled_Xest=scaler.fit_transform(X_test)
```

```
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression();
logreg.fit(rescaled_Xtrain,Y_train)
```

```
from sklearn.metrics import confusion_matrix
y_pred=logreg.predict(rescaled_Xest)
print("Accuarcy=",logreg.score(rescaled_Xest,Y_test))
print("confusion_matrix=",confusion_matrix(Y_test,y_pred))
```
